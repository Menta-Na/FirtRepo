```python
from string import ascii_letters

class Person:
    S_RUS = 'абвгдеёжзийклмнопрстуфхцчшщьыъэюя'
    S_RUS_UPPER = S_RUS.upper()
    
    def __init__(self, fio, old, ps, weight):
        self.verify_fio(fio)
        self.verify_old(old)
        self.verify_ps(ps)
        self.verify_weight(weight)
        
        self.__fio = fio.split()
        self.__old = old
        self.__passport = ps
        self.__weight = weight
        
    @classmethod
    def verify_fio(cls, fio):
        # Проверка типа и формата ФИО
        if type(fio) != str:
            raise TypeError("ФИО должно быть строкой")
        f = fio.split()
        if len(f) != 3:
            raise TypeError("Неверный формат записи")
        
        letters = ascii_letters + cls.S_RUS + cls.S_RUS_UPPER
        for s in f:
            if len(s) < 1:
                raise TypeError("В ФИО должен быть хотя бы один символ")
            if len(s.strip(letters)) != 0:
                raise TypeError("В ФИО можно использовать только буквы и дефис")
    
    @classmethod
    def verify_old(cls, old):
        # Проверка возраста
        if type(old) != int or old < 14 or old > 120:
            raise TypeError("Возраст должен быть целым числом в диапазоне [14; 120]")
            
    @classmethod
    def verify_weight(cls, w):
        # Проверка веса
        if type(w) != float or w < 20:
            raise TypeError("Вес должен быть вещественным числом от 20 и выше")
            
    @classmethod
    def verify_ps(cls, ps):
        # Проверка паспорта
        if type(ps) != str:
            raise TypeError("Паспорт должен быть строкой")
        s = ps.split()
        if len(s) != 2 or len(s[0]) != 4 or len(s[1]) != 6:
            raise TypeError("Неверный формат паспорта")
        for p in s:
            if not p.isdigit():
                raise TypeError("Серия и номер паспорта должны быть числами")

    @property
    def fio(self):
        return self.__fio 
    
    @property
    def old(self):
        return self.__old    
                
    @old.setter
    def old(self, old):
        self.verify_old(old)
        self.__old = old
        
    @property
    def weight(self):
        return self.__weight
    
    @weight.setter
    def weight(self, weight):
        self.verify_weight(weight)
        self.__weight = weight
        
    @property
    def passport(self):
        return self.__passport
        
    @passport.setter
    def passport(self, ps):
        self.verify_ps(ps)
        self.__passport = ps

# Создание объекта класса Person
p = Person("Заровский Сергей Алексеевич", 21, "1234 567890", 72.1)   
# Изменение возраста объекта
p.old  = 31
# Изменение паспортных данных объекта
p.passport = '4512 720394'
# Изменение веса объекта
p.weight = 70.5

# Вывод словаря атрибутов объекта для проверки изменений
print(p.__dict__)
```

```python
"""В этом коде свойства property используются для обеспечения контроля над доступом к атрибутам объекта класса 
Person. 
Вот как они используются:

fio, old, weight, passport: Эти свойства позволяют получить доступ к приватным 
    атрибутам объекта __fio, __old, __weight и __passport, соответственно,
    извне класса. Это делает эти атрибуты доступными для чтения извне, 
    но не для прямого изменения.
old, weight, passport: Эти свойства также имеют методы-сеттеры (old.setter, weight.setter, passport.setter),
    которые позволяют устанавливать новые значения для соответствующих атрибутов объекта с дополнительной
    проверкой на корректность нового значения.
Таким образом, использование свойств property обеспечивает контроль над доступом к данным объекта класса
и позволяет выполнять проверку и обработку данных при их чтении и изменении"""
```

```python
from string import ascii_letters

class Person:
    # Константы для русских букв
    S_RUS = 'абвгдеёжзийклмнопрстуфхцчшщьыъэюя'
    S_RUS_UPPER = S_RUS.upper()
    
    def __init__(self, fio, old, ps, weight):
        # Вызываем методы проверки данных для ФИО, возраста, паспорта и веса
        self.verify_fio(fio)
        self.verify_old(old)
        self.verify_ps(ps)
        self.verify_weight(weight)
        
        # Разбиваем ФИО на составляющие и сохраняем их
        self.__fio = fio.split()
        # Сохраняем возраст
        self.__old = old
        # Сохраняем паспортные данные
        self.__passport = ps
        # Сохраняем вес
        self.__weight = weight
        
    @classmethod
    def verify_fio(cls, fio):
        # Проверка типа и формата ФИО
        if type(fio) != str:
            raise TypeError("ФИО должно быть строкой")
        f = fio.split()
        if len(f) != 3:
            raise TypeError("Неверный формат записи")
        
        letters = ascii_letters + cls.S_RUS + cls.S_RUS_UPPER
        for s in f:
            if len(s) < 1:
                raise TypeError("В ФИО должен быть хотя бы один символ")
            if len(s.strip(letters)) != 0:
                raise TypeError("В ФИО можно использовать только буквы и дефис")
    
    @classmethod
    def verify_old(cls, old):
        # Проверка возраста
        if type(old) != int or old < 14 or old > 120:
            raise TypeError("Возраст должен быть целым числом в диапазоне [14; 120]")
            
    @classmethod
    def verify_weight(cls, w):
        # Проверка веса
        if type(w) != float or w < 20:
            raise TypeError("Вес должен быть вещественным числом от 20 и выше")
            
    @classmethod
    def verify_ps(cls, ps):
        # Проверка паспорта
        if type(ps) != str:
            raise TypeError("Паспорт должен быть строкой")
        s = ps.split()
        if len(s) != 2 or len(s[0]) != 4 or len(s[1]) != 6:
            raise TypeError("Неверный формат паспорта")
        for p in s:
            if not p.isdigit():
                raise TypeError("Серия и номер паспорта должны быть числами")

    @property
    def fio(self):
        # Получение ФИО
        return self.__fio 
    
    @property
    def old(self):
        # Получение возраста
        return self.__old    
                
    @old.setter
    def old(self, old):
        # Установка нового значения возраста с проверкой
        self.verify_old(old)
        self.__old = old
        
    @property
    def weight(self):
        # Получение веса
        return self.__weight
    
    @weight.setter
    def weight(self, weight):
        # Установка нового значения веса с проверкой
        self.verify_weight(weight)
        self.__weight = weight
        
    @property
    def passport(self):
        # Получение паспортных данных
        return self.__passport
        
    @passport.setter
    def passport(self, ps):
        # Установка новых паспортных данных с проверкой
        self.verify_ps(ps)
        self.__passport = ps

# Создание объекта класса Person с передачей данных для инициализации
p = Person("Заровский Сергей Алексеевич", 31, "1234 567890", 72)   
# Изменение возраста объекта
p.old  = 21
# Изменение паспортных данных объекта
p.passport = '4512 720394'
# Изменение веса объекта
p.weight = 70.5

# Вывод словаря атрибутов объекта для проверки изменений
print(p.__dict__)
```
